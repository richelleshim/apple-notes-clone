```python
import random

def generate_balanced_function(n):
  '''
  This function generates a balanced function where half are 0.
  In the best case, the f(0) != f(1). However, in the worst case,
  we would have to check ((2^n)/2) + 1 of the inputs to determine
  whether the function is balanced or not. The random.shuffle() function
  is used to create different orderings of a balanced function's outcome.
  '''
  inputs = list(range(2 ** n))
  random.shuffle(inputs)

  half = len(inputs) // 2
  halfZeroes = set(inputs[:half])
  def balanced_function(x):
    return 0 if x in halfZeroes else 1
  return balanced_function

def constant_zero_function(x):
  '''
  A constant function that always returns 0
  '''
  return 0

def constant_one_function(x):
  '''
  A constant function that always returns 1
  '''
  return 1

def input_function(n):
  '''
  This is to generate my "black box" function, where
  I don't know what function is being used. Determined by
  a pseudorandom integer generated by the randint() function,
  the functions are generated as either a balanced, constant zero,
  or constant one function.
  '''
  option = random.randint(0, 2)
  if option == 0:
    return generate_balanced_function(n)
  elif option == 1:
    return constant_zero_function
  else:
    return constant_one_function


def deutsch_jozsa(f, n):
  '''
  If the function is balanced, then there are both 1s and 0s.
  If the function is constant, then there is only one value.
  We compare f(0) until we either find a f(x) != f(0) or
  until we have checked (2^n / 2) + 1 outputs.
  If all (2^n / 2) + 1 outputs are the same, then the function is constant.
  If not, the function is balanced.
  '''
  firstOutput = f(0)
  for x in range((((2 ** n)//2)+1)):
    if f(x) != firstOutput:
      return "Balanced"
  return "Constant"



#testing my code:

n = 6
testFunction = input_function(n) #this can be any of the 3 functions

#this tells us the result of the deutsch_jozsa function:
result = deutsch_jozsa(testFunction, n)
print(f"Function is: {result}")

#this prints out the outputs of the function
#to make checking balanced functions easier, I added calculate and printed the sum

print("Outputs:")
sumOfOnes = 0
for i in range(2 ** n):
  print(f"f({i}) = {testFunction(i)}")
  sumOfOnes += testFunction(i)
print(sumOfOnes)

#test cases:

#for balanced function

n = 5
balancedFunction = generate_balanced_function(n)
resultFunction = deutsch_jozsa(balancedFunction, n)
print(f"Function is: {resultFunction}")
for i in range(2 ** n):
  print(f"f({i}) = {balancedFunction(i)}")


#for constant functions

n = 5
constantFunction = constant_zero_function
resultFunction = deutsch_jozsa(constantFunction, n)
print(f"Function is: {resultFunction}")
for i in range(2 ** n):
  print(f"f({i}) = {constantFunction(i)}")

n = 5
constantOnesFunction = constant_one_function
resultFunction = deutsch_jozsa(constantOnesFunction, n)
print(f"Function is: {resultFunction}")
for i in range(2 ** n):
  print(f"f({i}) = {constantOnesFunction(i)}")
```
